```python
income_order = ["$120K +", "$80K - $120K", "$60K - $80K", "Less than $40K", "Unknown"]
df['Income_Category'] = pd.Categorical(df['Income_Category'], categories=income_order, ordered=True)

# Crear el gráfico de distribución por género y categoría de ingresos
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Income_Category', hue='Gender', palette='viridis')
plt.title('Distribución de Género por Categoría de Ingresos')
plt.xlabel('Categoría de Ingresos')
plt.ylabel('Número de Clientes')
plt.legend(title='Género')
plt.xticks(rotation=45)
plt.show()
```

```python
credit_limit_stats = round(df.groupby(['Income_Category', 'Gender'])['Credit_Limit'].mean().unstack(),2)
credit_limit_stats
```

Se observa que el limite de credito es mayor en el genero Masculino igual debido a que son los que tienen mayor ingresos

```python
plt.figure(figsize=(12, 6))
credit_limit_stats.plot(kind='bar', ax=plt.gca())
plt.title('Promedio del Límite de Crédito según Categoría de Ingresos y Género')
plt.xlabel('Categoría de Ingresos')
plt.ylabel('Promedio del Límite de Crédito')
plt.legend(title='Género')
plt.xticks(rotation=45)
plt.show()
```

```python
education_order = [
    "Doctorate", "Post-Graduate", "College", "Graduate", "High School", "Uneducated", "Unknown"
]
df['Education_Level'] = pd.Categorical(df['Education_Level'], categories=education_order, ordered=True)

income_education_stats = df.groupby(['Education_Level', 'Income_Category']).size().unstack(fill_value=0)
income_education_stats
```
